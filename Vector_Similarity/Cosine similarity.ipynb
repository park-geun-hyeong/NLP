## Cosine Similarity

- 두 벡터 간의 코사인 각도를 이용하여 구할 수 있는 두 벡터의 유사도
- 두 벡터가 가리키는 방향이 얼마나 유사한지

```python
import pandas as pd
import numpy as np
import os

path = os.getcwd() 
```

```python
movie = pd.read_csv(path + '/movies_metadata.csv/movies_metadata.csv')  
```

```python
movie.shape
```

```python
movie.isna().sum()
```

```python
data = movie.head(20000)
data['overview']
```

```python
data['overview'] = data['overview'].fillna('')  
```

```python
data.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

```python
tf = TfidfVectorizer().fit(data['overview'])
tf_mat = tf.transform(data['overview'])   
```

```python
tf_mat.toarray().shape
```

```python
tf.vocabulary_
```

### tfidf를 사용하여 각 영화에대한 cos 유사도 구하기

```python
cos_sim = linear_kernel(tf_mat, tf_mat)
```

```python
pd.DataFrame(cos_sim)
```

```python
def recomandation(title, cos_sim = cos_sim):
    indices = pd.Series(data.index, index = data['title']).drop_duplicates()
    idx = indices[title]
    sim_scores = [i for i in enumerate(cos_sim[idx])]
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_movie_score = sim_scores[1:11]
    sim_movie_idx = [i[0] for i in sim_movie_score]
    
    sim_movie_title = data['title'].loc[sim_movie_idx]
    return sim_movie_title
```

```python
recomandation('The Dark Knight Rises')
```
